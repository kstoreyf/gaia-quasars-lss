```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from astropy.table import Table
```

```python
tmpz = Table.read('../data/twompz.fits.gz')
tmpz.columns
```

```python
np.sum(tmpz['ZSPEC'] == -999.0) / len(tmpz['ZSPEC'])
```

```python
idx_lowz = (tmpz['ZPHOTO'] > 0.05) & (tmpz['ZPHOTO'] < 0.1)
idx_lowz = idx_lowz.flatten() # dont know why have to do this for 2mpz, but do
tmpz_lowz = tmpz[idx_lowz]
print(len(tmpz), len(tmpz_lowz))
```

```python
plt.figure(figsize=(9,6))
plt.scatter(tmpz_lowz['L'], tmpz_lowz['B'], s=0.1, alpha=0.1, c='k')
plt.xlabel('')
plt.ylabel('b')
```
